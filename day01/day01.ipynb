```python
from pathlib import Path
file = open(Path("input.txt"), 'r')

lines = file.readlines()

temp_cals = 0
cal_list = []

for line in lines:
    line = line.strip('\n')
    if (line == ''):
        cal_list.append(temp_cals)
        print("this elves calories: " + str(temp_cals))
        temp_cals = 0
    else:
        temp_cals += int(line)
print("max calories: " + str(max(cal_list)))

cal_list.sort(reverse=True)

import numpy as np

x=3

xSum = np.sum(np.array(cal_list)[:x])
print("Sum of x highest: " + str(xSum))
```
