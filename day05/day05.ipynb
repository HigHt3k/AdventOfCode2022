```python
from pathlib import Path

file = open(Path("input.txt"), 'r')

# define crates strings
crates = [next(file).strip("\n") for x in range(8)]
next(file)
next(file)

# define movement strings
movements = file.read().splitlines()

# parse crates correctly

def parse_crates(c, skip_every_n):
    n = skip_every_n
    crates_list = []
    for c_line in c:
        # remove every 4th character
        s_new = c_line[0:3]
        s_add = c_line[3:]
        for i in range(0, len(s_add)):
            if i % 4 != 0:
                s_new += s_add[i]
        # split every 3 chars
        crates_list.append([s_new[i:i+n] for i in range(0, len(s_new), n)])
    return crates_list

#  parse movements
def parse_movements(m):
    m_list = []
    for movement in m:
        times = movement.split("move ")[1].split(" from")[0]
        from_pos = movement.split("from ")[1].split(" to")[0]
        to_pos = movement.split("to ")[1]
        packed = [int(times), [int(from_pos)-1, int(to_pos)-1]]
        m_list.append(packed)
    return m_list

# find highest crate
def highest_at_nth_col(crate_status, nth_col):
    row = 0
    for c_row in crate_status:
        if c_row[nth_col] != '   ':
            return [row,nth_col]
        row+=1
    return [row, nth_col]

# define do move
# movement single: [from, to]
def move(movement, crate_status):
    from_pos = movement[0]
    to_pos = movement[1]
    highest_crate_from = highest_at_nth_col(crate_status, from_pos)
    highest_crate_to = highest_at_nth_col(crate_status, to_pos)
    crate_element = crate_status[highest_crate_from[0]][highest_crate_from[1]]
    # set old element to free
    crate_status[highest_crate_from[0]][highest_crate_from[1]] = '   '
    # set new crate to position
    if highest_crate_to[0] >= 1:
        crate_status[highest_crate_to[0] - 1][highest_crate_to[1]] = crate_element
    else:
        crate_status.insert(0, ['   ' for x in range(0, 9)])
        crate_status[highest_crate_to[0]][highest_crate_to[1]] = crate_element
    return crate_status

crate_list = parse_crates(c = crates, skip_every_n=3)
movement_list = parse_movements(m = movements)

assert (highest_at_nth_col(crate_list, 8)) == [0, 8]
assert (highest_at_nth_col(crate_list, 3)) == [3, 3]

# do movements
for movement in movement_list:
   for i in range(0, movement[0]):
       m = movement[1]
       crate_list = move(m, crate_list)

def print_highest_object(c):
    highest = ['','','','','','','','','']
    for line in c:
        col_count = 0
        for col in line:
            if col != '   ':
                if highest[col_count] == '':
                    highest[col_count] = col
            col_count += 1
    final = ''
    for h in highest:
        final += h.strip("[").strip("]")
    print(final)

print_highest_object(crate_list)
```

```python
# define multi move method
def multi_move(count, movement, crate_status):
    from_pos = movement[0]
    to_pos = movement[1]

    highest_crate_to = highest_at_nth_col(crate_status, to_pos)

    crate_elements = []

        # set new crate to position
    diff = highest_crate_to[0] - count
    if diff <= 0:
        for i in range(0, abs(diff)):
            crate_status.insert(0, ['   ' for x in range(0, 9)])

    highest_crate_from = highest_at_nth_col(crate_status, from_pos)
    for c in range(0, count):
        crate_elements.append(crate_status[highest_crate_from[0] + c][highest_crate_from[1]])
        crate_status[highest_crate_from[0] + c][highest_crate_from[1]] = '   '

    highest_crate_to = highest_at_nth_col(crate_status, to_pos)

    for c in range(0, count):
        crate_status[highest_crate_to[0] - count + c][highest_crate_to[1]] = crate_elements[c]

    return crate_status

crate_list = parse_crates(c = crates, skip_every_n=3)
movement_list = parse_movements(m = movements)

assert (highest_at_nth_col(crate_list, 8)) == [0, 8]
assert (highest_at_nth_col(crate_list, 3)) == [3, 3]

def print_self(c):
    for c_ in c:
        print(c_)

print_self(crate_list)

for movement in movement_list:
   crate_list = multi_move(movement[0], movement[1], crate_list)
   print(movement)
   print_self(crate_list)

print_highest_object(crate_list)

```
