```python
from pathlib import Path

file = open(Path("input.txt"), "r")

#set True for part 1, False for part 2
worry = True

# parse and apply monkey operation
def operation(operation, item):
    if "*" in operation:
        if "old * old" in operation:
            item *= item
        else:
            by = int(operation.split("* ")[1])
            item *= by
    elif "+" in operation:
        by = int(operation.split("+ ")[1])
        item += by
    else:
        print("Wrong operation found")
        exit(-1)
    return item

# do the test to identify action of monkey
def test(item, divide_by):
    if item % divide_by == 0:
        return True
    else:
        return False


# monkey business :)
class Monkey:
    def __init__(self, id, items, operation, test_division,
                 on_true, on_false, worry):
        self.id = id
        self.items = items
        self.operation = operation
        self.test_division = test_division
        self.on_true = on_true
        self.on_false = on_false
        self.inspections = 0
        self.worry = worry

    def add_item(self, item):
        self.items.append(item)

    def get_id(self):
        return self.id

    def process(self):
        for i in range(0, len(self.items)):
            self.items[i] = operation(self.operation, self.items[i])
            self.inspections += 1
            if not self.worry:
                print("Worry level is declining")
                self.items[i] = int(self.items[i]/3)
            else:
                self.items[i] = int(self.items[i]%div)
            test_val = test(self.items[i], self.test_division)

            if test_val:
                throw_to_monkey(self.items[i], self.on_true)
            else:
                throw_to_monkey(self.items[i], self.on_false)
        self.items = []

    def print_self(self):
        print(self.id, ":", self.items, "\n\top:", self.operation, "\n\ttest:", self.test_division, self.on_true, self.on_false, "\n\tTotal inspections:", self.inspections)

def throw_to_monkey(item, to):
    for m in monkeys:
        if m.get_id() == to:
            m.add_item(item)
            break
# parse file
def monkey_parser(file):
    monkey_list = []
    lines = file.read().splitlines()
    for i in range(0, len(lines), 7):
        id = int(lines[i].split("Monkey ")[1].strip(":"))
        items = lines[i+1].split("items:")[1].split(",")
        item_list = [int(x) for x in items]
        operation = lines[i+2].split("Operation: ")[1]
        test = int(lines[i+3].split("Test: divisible by ")[1])
        on_true = int(lines[i+4].split("monkey ")[1])
        on_false = int(lines[i+5].split("monkey ")[1])
        monkey_list.append(Monkey(id, item_list, operation, test, on_true, on_false, worry))
    return monkey_list

def print_all_monkeys(monkeys):
    for m in monkeys:
        m.print_self()

def find_divisor(monkeys):
    div = 1
    for monkey in monkeys:
        div *= monkey.test_division
    return div

monkeys = monkey_parser(file)
div = find_divisor(monkeys)
print(div)
# print("Round: 0")
# print_all_monkeys(monkeys)
# simulate n rounds
n = 10000
rounds = [i for i in range(1, n+1)]

for round in rounds:
    # print("Round:", round)
    print("calculating monkey business:", round/len(rounds)*100, "%")
    for monkey in monkeys:
        monkey.process()
    # print_all_monkeys(monkeys)

# find two highest monkeys
inspection_counts = []
for monkey in monkeys:
    inspection_counts.append(monkey.inspections)

inspection_counts.sort(reverse=True)
print(inspection_counts)
print("Monkey Business Factor:",inspection_counts[0] * inspection_counts[1])
```
