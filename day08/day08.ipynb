```python
from pathlib import Path
from copy import deepcopy

# read file
file = open(Path("input.txt"), 'r')
forest = file.read().split("\n")

for i in range(0, len(forest)):
    forest[i] = [*forest[i]]

for i in range(0, len(forest)):
    for j in range(0, len(forest[i])):
        forest[i][j] = int(forest[i][j])

def check_left(forest_input):
    forest_render = deepcopy(forest_input)
    for y in range(0, len(forest)):
        highest = -1
        for x in range(0, len(forest[y])):
            if forest[y][x] > highest:
                highest = forest[y][x]
                forest_render[y][x] = "+"
    return forest_render

def check_right(forest_input):
    forest_render = deepcopy(forest_input)
    # check from left side view
    for y in range(0, len(forest)):
        highest = -1
        for x in range(len(forest[y])-1, 0, -1):
            if forest[y][x] > highest:
                highest = forest[y][x]
                forest_render[y][x] = "+"
    return forest_render

def check_top(forest_input):
    forest_render = deepcopy(forest_input)
    # check from left side view
    for x in range(0, len(forest[0])):
        highest = -1
        for y in range(0, len(forest)):
            if forest[y][x] > highest:
                highest = forest[y][x]
                forest_render[y][x] = "+"
    return forest_render

def check_bottom(forest_input):
    forest_render = deepcopy(forest_input)
    # check from bottom side view
    for x in range(0, len(forest[0])):
        highest = -1
        for y in range(len(forest)-1, 0, -1):
            if forest[y][x] > highest:
                highest = forest[y][x]
                forest_render[y][x] = "+"
    return forest_render

bottom = check_bottom(forest_input=forest)
left = check_left(forest_input=forest)
top = check_top(forest_input=forest)
right = check_right(forest_input=forest)

def overwrite(original, overwrite):
    for i in range(0, len(original)):
        for j in range(0, len(original[i])):
            if overwrite[i][j] == "+":
                original[i][j] = "+"
            else:
                pass

def calc_result(res, print=False):
    count_trees = 0
    for i in result:
        s = ""
        for j in i:
            s += str(j)
            if j == "+":
                count_trees += 1
        if print:
            print(s)
    return count_trees

result = deepcopy(bottom)
overwrite(result, bottom)
overwrite(result, left)
overwrite(result, top)
overwrite(result, right)
tree_count = calc_result(result, print=False)
print("number of trees that are visible: ", tree_count)

def tree_count(vals, cur_height):
    distance = 0
    for v in vals:
        if v >= cur_height:
            return distance+1
        distance += 1
    return distance

def count_trees_left(trees, x, y):
    relevant_vals = trees[y][:x]
    relevant_vals.reverse()
    return tree_count(relevant_vals, trees[y][x])

def count_trees_right(trees, x, y):
    relevant_vals = trees[y][x+1:]
    return tree_count(relevant_vals, trees[y][x])

def count_trees_down(trees, x, y):
    relevant_vals = [trees[i][x] for i in range(y+1, len(trees))]
    return tree_count(relevant_vals, trees[y][x])

def count_trees_up(trees, x, y):
    relevant_vals = [trees[i][x] for i in range(0, y)]
    relevant_vals.reverse()
    return tree_count(relevant_vals, trees[y][x])

def calc_scenic_view(trees):
    best_scenic_view = 0
    for y in range(0, len(trees)):
        for x in range(0, len(trees[y])):
            scenic_view = count_trees_down(trees, x, y) * count_trees_up(trees, x, y) * count_trees_left(trees, x, y) * count_trees_right(trees, x, y)
            if scenic_view > best_scenic_view:
                best_scenic_view = scenic_view
    return best_scenic_view

print("The best scenic view is: ", str(calc_scenic_view(trees=forest)))
```
